Temporary cleaning script to remove images with index names

```python
import os
ipc_500_root = './CIFAR10_DM/ipc500'
classes = os.listdir(ipc_500_root)
```

```python
for cls in classes:
    cls_dir = os.path.join(ipc_500_root, cls)
    for image in os.listdir(cls_dir):
        # Remove indexed images, eg. 100.png
        if len(image) < 10: 
            os.remove(os.path.join(cls_dir, image))
```

Check for number of images per class

```python
[len(os.listdir(os.path.join(ipc_500_root, x))) for x in classes]
```

512 images per class for ipc 500

Generate index file for these images

```python
class_to_idx = {'airplane': 0,
                'automobile': 1,
                'bird': 2,
                'cat': 3,
                'deer': 4,
                'dog': 5,
                'frog': 6,
                'horse': 7,
                'ship': 8,
                'truck': 9}
```

```python
import pandas as pd

entries = []
for cls in classes:
    cls_idx = class_to_idx[cls]
    cls_dir = os.path.join(ipc_500_root, cls)
    for image in os.listdir(cls_dir):
        # On windows, manually writing path in linux format
        entries.append([f'./CIFAR10_DM/ipc500/{cls}/{image}', cls_idx])
entries
```

```python
df = pd.DataFrame(entries, columns=['image_path', 'label'])
df
```
